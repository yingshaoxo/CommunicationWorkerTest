```python
import re
from pprint import pprint
```

```python
def printlist(l):
    for e in l:
        pprint(e)
        print('-'*12)
```

```python
with open("1.txt", 'r') as f:
    question = f.read()
print(question[:200])
```

```python
question_list = re.split(r'\s\d+\. ', question) 
print(f"parts included: {len(question_list)}\n")

question_list = [question for question in question_list if "答案" in question]

printlist(question_list)
```

```python
#    {
#      "id": 1,
#      "title": "驾驶技能准考证明的有效期是多久？",
#      "items": [
#        "1年",
#        "2年",
#        "3年",
#        "4年"
#      ],
#      "answer": 2,
#      "reason": "驾驶技能准考证明的有效期为三年，申请人应当在有效期内完成科目二和科目三考试。未在有效期内完成考试的，已考试合格的科目成绩作废。"
#    },

data_list = []
item = {}
for index, string in enumerate(question_list):
    item["id"] = index
    
    sentences = string.split("\n")
    sentences = [sentences[0]] + [sentence.strip() for sentence in sentences[1:] if "○" in sentence or "□" in sentence or "答案" in sentence]
    
    item["question"] = sentences[0]
    item["choices"] = "\n".join(sentences[1:-1])
    if "答案：" in item['choices']:
        continue
    
    answers = list(sentences[-1].replace("答案：", "").lower())
    item["answer"] = ''.join(answers)
    
    data_list.append(item.copy())

printlist(data_list)
```
