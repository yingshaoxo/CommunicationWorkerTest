```python
import pandas as pd

xls = pd.ExcelFile("铁路通信工中级(线务维护).xls")
sheet = xls.parse(0) #2 is the sheet number
```

```python
def clear(dataframe):
    _list = dataframe.tolist()
    new_list = []
    for text in _list:
        text = str(text)
        text = text.strip(' \n')
        if text == 'nan':
            text = ""
        new_list.append(text)
    return new_list

title = clear(sheet['试题内容'])
answer = clear(sheet['正确选项'])
A = clear(sheet['A'])
B = clear(sheet['B'])
C = clear(sheet['C'])
D = clear(sheet['D'])
answer_index = []
for i in answer:
    if i == 'A':
        answer_index.append(0)
    elif i == 'B':
        answer_index.append(1)
    elif i == 'C':
        answer_index.append(2)
    elif i == 'D':
        answer_index.append(3)
```

```python
data_list = []

id_ = 0
for title_, answer_index_, A_, B_, C_, D_ in zip(title, answer_index, A, B, C, D):
    id_ += 1
    print(id_)
    print(title_)
    print(answer_index_)
    print(A_)
    print(B_)
    print(C_)
    print(D_)
    print('-' * 12)
    one = {
        'id': id_,
        'title': title_,
        'items': [
            A_,
            B_,
            C_,
            D_
        ],
        'answer': answer_index_,
        'reason': ''
    }
    data_list.append(one) 
```

```python
final_data = {'list': data_list}
```

```python
import json

text = json.dumps(final_data)

with open('data.json', 'w', encoding='utf-8') as f:
    f.write(text)
```
